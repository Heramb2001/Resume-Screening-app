```python
import re
```

```python
def clean(data):
    cleantext=re.sub('http\S+\s',' ',data)
    cleantext=re.sub('@\S+',' ',cleantext)
    cleantext=re.sub('#\S+',' ',cleantext)
    cleantext=re.sub('â¢ r â¢',' ',cleantext)
    cleantext=re.sub('â¢',' ',cleantext)
    cleantext=re.sub('\\r\\n',' ',cleantext)
    cleantext=re.sub('â',' ',cleantext)
    
    cleantext=re.sub('/ \d{4} /',' ',cleantext)
    return cleantext
```

```python
df['Resume']=df['Resume'].apply(clean)
```

```python
def removespecial(text):
    x=''
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
    
```

```python
df['Resume']=df['Resume'].apply(removespecial)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    x=[]#Empty list
    for i in text.split(): #text to  list conversion
        if i not in stopwords.words('english'):
            x.append(i) #append in the empty list
    y=x[:] #copying all from x to y so that next time x will be empty list
    x.clear()
    return y
```

```python
df['Resume']=df['Resume'].apply(remove_stopwords)
```

```python
remove_stopwords('Hello my namee is and you are the greatest/')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def portstemer(text):
    y=[]
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
portstemer(['I','Loving'])
```

```python
df['Resume']=df['Resume'].apply(portstemer)
```

```python
df['Resume']
```

```python
def join_back(list_input): 
    return " ".join(list_input)
```

```python
df['Resume']=df['Resume'].apply(join_back)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming df is your DataFrame and 'Resume' is the column containing text data
tfidf = TfidfVectorizer(stop_words='english')

# Fit and transform the text data
tfidf_matrix = tfidf.fit_transform(df['Resume'])

# Get the number of samples (rows) in the sparse matrix
num_samples = tfidf_matrix.shape[0]

# Now you can use num_samples or any other attribute/method as needed
print(f'Number of samples: {num_samples}')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(tfidf_matrix,df['Category'],test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier 
from sklearn.metrics import accuracy_score
clf=OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print(y_pred)
print(accuracy_score(y_test,y_pred))
```
